# WeatherPy
----
Using a randomly generated sample of 549 cities at latitutudes between +/- 90 degrees and longitudes +/- 180 degrees, I examined the relationships between maximum temperature, humidity, cloudiness, and wind speed and latitude to test the hypothesis that it gets warmer the closer one gets to the equator (0 degrees latitude).

### Analysis
1. The scatterplot of latitude vs Temperature(F), shows the relationship between temperature and latitude.  As espected, the closer one gets to 0 degrees latitude the warmer it gets.   The further away both in positive and negative directions, the cooler it gets.

2. While there is no relationship between humidity and latitude, few cities are located in places with humidity below 40%.

3. While there is not relationship between humidity and latitude, few cities are located in places with wind speeds above 25 mph.



```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
from pprint import pprint
import csv

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

### Perform API Calls
* Perform a weather check on each city using a series of successive API calls.
* Include a print log of each city as it'sbeing processed (with the city number and city name).


```python
url = " http://api.openweathermap.org/data/2.5/weather?q="
api_key = "&APPID=666c4dcbc13d170d764b8ea8089a36a0"
unit ="&units=imperial"
sample =[]
n = 0
for city in cities:
    data = requests.get(url + city + api_key + unit).json()
    sample.append(data)
    if data["cod"] == "404":
        print("city not found")
    else:
        n += 1
        print("City number " + str(n) + " " + data['name'])
       
    
```

```python
pprint(sample)
```

```python
city = []
cloud = []
country = []
date =[]
humidity = []
lat =[]
lng = []
maxtemp = []
windspeed = [] 


for item in sample:
    try:
        city.append(item["name"])
        cloud.append(item["clouds"]["all"])
        country.append(item["sys"]["country"])
        date.append(item["dt"])
        humidity.append(item["main"]["humidity"])
        lat.append(item["coord"]["lat"])
        lng.append(item["coord"]["lon"])
        maxtemp.append(item["main"]["temp_max"])
        windspeed.append(item["wind"]["speed"])
    except KeyError:
        continue
    
```

```python
df= pd.DataFrame({"City": city, 
                  "Country": country, 
                  "Date": date, 
                  "Latitude": lat, 
                  "Longitude":lng,
                  "Cloudiness": cloud, 
                 "Humidity": humidity,
                 "Max Temp": maxtemp,
                 "Wind Speed": windspeed})
```

#### Latitude vs. Temperature Plot

```python
latvtemp = plt.scatter(df["Latitude"], df["Max Temp"], c = "blue", edgecolor = "black" )
plt.title("Temperature (F) vs. Latitude (8/31/2018)")
plt.xlabel("Latitude")
plt.ylabel("Temperature (F)")
plt.tight_layout()
plt.savefig("lat_v_temp.png")
```

#### Latitude vs. Humidity Plot

```python
latvhumid = plt.scatter(df["Latitude"], df["Humidity"], c = "blue", edgecolor = "black" )
plt.title("Humidity (%) vs. Latitude (8/31/2018)")
plt.xlabel("Latitude")
plt.ylabel("Humidity (%)")
plt.tight_layout()
plt.savefig("lat_v_humid.png")
```

#### Latitude vs. Cloudiness Plot

```python
latvcloud = plt.scatter(df["Latitude"], df["Cloudiness"], c = "blue", edgecolor = "black" )
plt.title("Cloudiness (%) vs. Latitude (8/31/2018)")
plt.xlabel("Latitude")
plt.ylabel("Clouddiness (%)")
plt.tight_layout()
plt.savefig("lat_v_cloud.png")
```

#### Latitude vs. Wind Speed Plot

```python
latvwind = plt.scatter(df["Latitude"], df["Wind Speed"], c = "blue", edgecolor = "black" )
plt.title("Wind Speed (mph) vs. Latitude (8/31/2018)")
plt.xlabel("Latitude")
plt.ylabel("Wind Speed (mph)")
plt.tight_layout()
plt.savefig("lat_v_wind.png")
```
